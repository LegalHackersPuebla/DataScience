```python
#Obtenemos el número de incidencias por cada categoría de la columna main_category.
df["main_category"].value_counts()
```

```python
#Con pandas "plot"

# Crea tu "canvas"
fig, ax = plt.subplots(figsize = (20,10))

#Damos formato
plt.style.use(['seaborn-whitegrid','seaborn-poster'])

#Creamos el plot

df["main_category"].value_counts().plot.bar()
```

```python
#Con matplotlib "plt"

# Crea tu "canvas"
fig, ax = plt.subplots(figsize = (20,10))

#Damos formato 
ax.set_xlabel('Este es mi eje X')
ax.set_ylabel('Este es mi eje Y')
ax.set_title(df["main_category"].name, size=35)
plt.xticks(rotation = 45, size=20) 

#Creamos el plot
index=list(df["main_category"].value_counts().index)
values=list(df["main_category"].value_counts())
plt.bar(index,values)
```

```python
#Obtenemos el número de incidencias por cada categoría de la columna currency.
df["currency"].value_counts()
```

```python
#Con matplotlib "plt"

# Crea tu "canvas"
fig, ax = plt.subplots(figsize = (20,10))

#Damos formato 
ax.set_xlabel('Este es mi eje X')
ax.set_ylabel('Este es mi eje Y')
ax.set_title(df["currency"].name, size=35)
plt.xticks(rotation = 30) 

#Creamos el plot
df["currency"].value_counts().plot.bar()
```

```python
#Con pandas "plot"

# Crea tu "canvas"
fig, ax = plt.subplots(figsize = (20,10))

#Damos formato
plt.style.use(['seaborn-whitegrid','seaborn-poster'])

#Creamos el plot
df["currency"].value_counts().plot.bar()
```

```python
#Con seaborn COUNTPLOT

_=sns.countplot(data=df, x="currency")
```

```python
#Con seaborn BARPLOT
frecuencia = df.currency.value_counts(normalize = True).values
categorias = df.currency.value_counts().index
_=sns.barplot(x=categorias,y=frecuencia)
```

```python
#Convertimos a Booleano una columna con np.where()
backers_bool=np.where(df["backers"]>0, 1,0)
df.insert(11, "backers_bool",backers_bool)
df
```

```python
#Ordenamos el data frame en función de la columna category
df.sort_values(by=["category"])
```

```python
#Ordenamos la columna backers ascendientemente
df["backers"].sort_values(ascending=True)
```

```python
#Verificamos si existen valores nulos en nuestro DataFrame para eliminarlos con dropnull()
df.isnull().sum()
```

```python
#Eliminamos registros con valores nulos y guardamos nuestro Dataframe
df.dropna(inplace=True)

#comprobamos
df.isnull().sum()
```

```python
#Verificamos si existen registros Duplicados
df.duplicated().sum()
```

```python
#En caso de haber duplicados los eliminamos
df.drop_duplicates(inplace=True)
df
```
